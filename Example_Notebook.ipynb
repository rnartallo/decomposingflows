```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial import Delaunay
import SimplicialComplexDecomp as SCD
import vectorfieldhelpers as VF
import importlib
import IsotropicOU as IOU
import MarkovChainHelpers as MC
```

# Generate an ensemble of stochastic trajectories and sub-sample

In this case we use the OU process

```python
theta = 5
no_trajectories = 200
samples=10000
tau=0.01
eps=0.5
Y =[]
trajs =[]
for i in range(0,no_trajectories):
    x0 = np.random.normal(0,1,2)
    X, X_rev, X_irrev, X_check = IOU.isotropic_OU2(theta,eps,x0,tau,samples)
    Y.append(X)
    trajs.append(X)
U =np.concatenate(Y, axis=1)
subUidx,subU  = VF.furthest_point_sampling(U.T,250)
subU = subU.T

sns.set_style('white')
sns.set_palette('viridis')
fig, axs = plt.subplots(1,1)
fig.set_size_inches((5,5))

palette = sns.color_palette("viridis", 1)
sns.scatterplot(x=subU[0,:], y=subU[1,:],ax=axs,c='black',linewidth = 0,s=3)

axs.set_title('Subsampled')

plt.tight_layout()
plt.show()
```

# Tesselate the subsampled points with the Delauney triangulation

```python
tesselation = Delaunay(subU.T)
vertices, edges, triangles = SCD.tesselation_to_simplicial_complex(subU.T,tesselation.simplices)
mid_points = VF.tesselation_mid_points(triangles,tesselation.points)
plt.triplot(subU.T[:,0], subU.T[:,1], tesselation.simplices,color='black',linewidth=0.5)
plt.scatter(x=mid_points[:,0],y=mid_points[:,1],color ='red',s=1)
plt.show()
```

# Estimate the transition intensities between bins

```python
T = MC.transition_rate_estimation(tesselation,Y)
```

# Take the dual triangulation

```python
tesselation2 = Delaunay(mid_points)
vertices, edges, triangles = SCD.tesselation_to_simplicial_complex(mid_points,tesselation2.simplices)
plt.triplot(mid_points[:,0], mid_points[:,1], tesselation2.simplices,color='black',linewidth=0.4)
```

# Define the edge-flow on the dual triangulation

```python
ef = MC.edge_flow_from_transition_matrix(T,edges)
Y = mid_points.T
cmap = plt.get_cmap('viridis')
normflow = plt.Normalize(0, np.max([ef.max()/2,-ef.min()/2]))
fig, ax= plt.subplots(1,1)
for i in range(0,len(edges)):
    e = edges[i]
    if ef[i]<0:
        ax.arrow(Y[0,e[1]],Y[1,e[1]],Y[0,e[0]]-Y[0,e[1]], Y[1,e[0]]-Y[1,e[1]],color = cmap(normflow(-ef[i])),head_width = 0.05)
    if ef[i]>0:
        ax.arrow(Y[0,e[0]],Y[1,e[0]],Y[0,e[1]]-Y[0,e[0]], Y[1,e[1]]-Y[1,e[0]],color = cmap(normflow(ef[i])),head_width = 0.05)
plt.show()
```

# Perform the discrete HHD using sparse LSQR

```python
X_curl, X_grad = SCD.helmholtz_decomposition_sparse_fast(ef,vertices,edges)
```

```python
C,G = SCD.proportion_of_flow_no_h(X_curl.reshape((len(edges))),X_grad.reshape((len(edges))),ef)
Y = mid_points.T
cmap = plt.get_cmap('viridis')
sns.set_style('white')
sns.set_palette('viridis')
fig, axs = plt.subplots(1,2)
fig.set_size_inches((8,4))

normflow = plt.Normalize(0, np.max([X_curl.max()/2,-X_curl.min()/2]))
for i in range(0,len(edges)):
    e = edges[i]
    if X_curl[i]<0:
        axs[0].arrow(Y[0,e[1]],Y[1,e[1]],Y[0,e[0]]-Y[0,e[1]], Y[1,e[0]]-Y[1,e[1]],color = cmap(normflow(-X_curl[i])),head_width = 0.05)
    if X_curl[i]>0:
        axs[0].arrow(Y[0,e[0]],Y[1,e[0]],Y[0,e[1]]-Y[0,e[0]], Y[1,e[1]]-Y[1,e[0]],color = cmap(normflow(X_curl[i])),head_width = 0.05)

normflow = plt.Normalize(0, np.max([X_grad.max()/2,-X_grad.min()/2]))

for i in range(0,len(edges)):
    e = edges[i]
    if X_grad[i]<0:
        axs[1].arrow(Y[0,e[1]],Y[1,e[1]],Y[0,e[0]]-Y[0,e[1]], Y[1,e[0]]-Y[1,e[1]],color = cmap(normflow(-X_grad[i])),head_width = 0.05)
    if X_grad[i]>0:
        axs[1].arrow(Y[0,e[0]],Y[1,e[0]],Y[0,e[1]]-Y[0,e[0]], Y[1,e[1]]-Y[1,e[0]],color = cmap(normflow(X_grad[i])),head_width = 0.05)
plt.show()
```
